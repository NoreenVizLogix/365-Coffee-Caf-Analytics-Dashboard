```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Input # Added Input
import pandas as pd
import numpy as np
```

```python
# Load the data
df_loaded = pd.read_csv('module_157_data.csv')   # Google Colab

# Separate features and target
X = df_loaded.drop('target', axis=1).values
y = df_loaded['target'].values

# RESHAPE: From (2000, 10) to (2000, 10, 1)
# This adds the "feature" dimension required by RNN
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Build and compile
model = Sequential([
    Input(shape=(10, 1)), # Explicitly define the input layer
    # SimpleRNN is the basic recurrent layer
    SimpleRNN(32), # input_shape is no longer needed here
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # (Adaptive Moment Estimation)
```

```python
# 1 # Train
model.fit(X, y, epochs=20, batch_size=32, validation_split=0.2)
```

```python
# 1 # Evaluate and print only the accuracy
print(f"\nFinal Accuracy: {model.evaluate(X, y, verbose=0)[1]:.4f}")
```
